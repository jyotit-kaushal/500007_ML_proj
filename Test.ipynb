# Principal Component Analysis

### Imports

```python
import pandas as pd
from sklearn.decomposition import PCA
```

### Data retrieval

```python
df = pd.read_csv("./Training and Testing sets/train_tfidf_features.csv")
```

### Reducing the dimensions

```python
label = df.loc[:,["label"]]
features_names = [str(i) for i in range(0,5000)]
features = df.loc[:,features_names]
```

##### 2000 features

```python
pca2000 = PCA(n_components=2000)
principalComponents2000 = pca.fit_transform(features)
principalDf2000 = pd.DataFrame(data = principalComponents2000)
principalDf2000
```

##### 1000 dimensions

```python
pca1000 = PCA(n_components=1000)
principalComponents1000 = pca1000.fit_transform(features)
principalDf1000 = pd.DataFrame(data = principalComponents1000)
principalDf1000
```

```python
pca500 = PCA(n_components=500)
principalComponents500 = pca500.fit_transform(features)
principalDf500 = pd.DataFrame(data = principalComponents500)
principalDf500
```

##### 200 dimensions

```python
pca200 = PCA(n_components=200)
principalComponents200 = pca200.fit_transform(features)
principalDf200 = pd.DataFrame(data = principalComponents200)
principalDf200
```

##### 100 dimensions

```python
pca100 = PCA(n_components=100)
principalComponents100 = pca100.fit_transform(features)
principalDf100 = pd.DataFrame(data = principalComponents100)
principalDf100
```
