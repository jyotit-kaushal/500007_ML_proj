```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('train_tfidf_features.csv')
df_train
```

```python
df_train_label = df_train.loc[:,"label"]
df_train_label = df_train_label.to_frame()
features = [str(i) for i in range(5000)]
df_train_features = df_train.loc[:, features]
```

```python
def normalize(X):
    X_out = X.copy()
    X_out = (X - X.mean(axis=0)) / X.std(axis=0)
    return X_out
    
def sigmoid(z):
    return (1/(1+np.exp(-z)))

def loss(y, y_hat):
    num_of_samples = len(y)
    loss = -(1/num_of_samples) * \
        np.sum(np.where(y==1, np.log(y_hat), np.log(1-y_hat)))
    return loss

def gradients(X, y, num_iters, lr):
    m, n = X.shape
    w = np.zeros((n,1))
    J_storage = np.zeros((num_iters,1))
    for i in range(num_iters):
        y_hat = sigmoid(np.matmul(X,w))
        derivative = (1/m) * np.matmul(X.T, (y_hat - y))
        w = w - lr * derivative 
        J_storage[i] = loss(y, y_hat) 
    return w, J_storage


def predict(X,w):
    x = normalize(X)
    preds = sigmoid((np.matmul(X, w)))
    return preds

def accuracy(y, y_hat):
    accuracy = np.sum(y == y_hat) / len(y)
    return accuracy
```

```python
train_w, training_loss = gradients(df_train_features, df_train_label, 100, 0.01)
```
